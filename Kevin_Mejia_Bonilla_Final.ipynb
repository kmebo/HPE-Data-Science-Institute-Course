```python
# Problem 1
#This code perform division with integer numbers which is a game for 3 to 4rd-grade students who
# are learning division
import random
def div(x=0):
    cnt = 0
    for i in range(x):
        a = random.randint(9, 1001)
        b = random.randint(1, 10)
        c = a // b
        d = a % b
        quotient_answer = int(input(f"{a}/{b}?\nEnter the Quotient: "))
        reminder_answer = int(input(f"Enter the Remainder: "))
        if quotient_answer == c and reminder_answer == d:
            print("Correct.")
            cnt += 1
        else:
            print("Incorrect.")
    print(f"You got {cnt} correct answers out of {x}")

print("Problem 1")
x= int(input("Provide an integer"))
div(x)
#Your function game() will take an integer n as input and then ask n  division questions. The numbers to be divided should be chosen randomly such that :
```

```python
import numpy as np
from numpy import *
import matplotlib
import matplotlib.pyplot as plt 
print("Problem 3\nThis code generate a series plot of the morning temperature in Fahrenheit vs time.")
try:
    #data = genfromtxt('stockholm_td_adj.dat')
    data = genfromtxt('stockholm_td_adj.dat')
    x = data.shape # get the number of rows and col

    # convert Morning values from Cel. to Fah.
    for i in range(x[0]):
        data[i][3] = round((data[i][3] * (9/5)) + 32, 1)

    # Afternoon Temperatures

    fig, ax = plt.subplots(figsize=(14,7))
    ax.plot(data[:,0] + data[:,1]/12.0 + data[:,2]/365, data[:,3])
    ax.axis('tight')
    ax.set_title('Morning Temperatures in Stockholm')
    ax.set_xlabel('Year')
    ax.set_ylabel('Temperature (F)');
except FileNotFoundError:
    print("File Not Found")
```
