```python
#Replace pass with body of statement to complete the homework
def split_the_check(number_of_friends):
    introduction = """  Kevin Mejia Bonilla, Peoplesoft ID: 1111111

     Program #4: Sharing the bill
     
     Scientific Computing With Python
     
     Fall 2023

     This program computes individual shares of a restaurant bill. \n"""
    print(introduction)
    names = []
    bills = []
    bill = 0
    for i in range(number_of_friends):
        names.append(input("Enter name of friend:"))
    print()
    for i in range(number_of_friends):
        x = input(f"Enter bill for {names[i]}:")
        bills.append(x)
        bill += float(x)
    name_str = "Names of Friends: "
    bill_str = "Individual bills: "
    for i in range(number_of_friends):
        name_str += names[i] + ", "
        bill_str += bills[i] + ", "
    
    name_str = name_str[0:len(name_str)-2]
    bill_str = bill_str[0:len(bill_str)-2]
    print("\n"+ name_str +"\n"+ bill_str +"\n")
    tip = input("Enter tip percentage:")
    
    total_bill = bill + (bill * (float(tip)/100))
    ind_bill = total_bill / number_of_friends
    print(f"Total bill plus tip: $ {total_bill:.2f} \nEach of us must pay $ {ind_bill:.2f}")
split_the_check(5)
```
