##Word count
Write a python program that takes a  text file name from the command line as input, reads in the file, and prints the number of occurrences of every word in the file. Your program should use a custom function for processing the number of occurrences, and this function should be case-insensitive so 'Hello' and 'hello' are treated as the same word. You should ignore words of length 2 or less. The results printed will be ordered from the most frequent to the least frequent. Hint: dictionary and list. Test your implementation on file great_expectations.txt.

```python
%%writefile wordcount.py 
import sys

#This function receives a string, line, and a dictionary, my_dict. my_dict. will be modified based on
# the information the string has.
def funct_1(line="", my_dict={}):
    #how to separate a string in a list
    temp_list = line.lower().split()
    for i in temp_list:
        if len(i) > 2:
            # In case, the word exists in the dictionary, just count it one more time
            if i in my_dict:
                my_dict[i] += 1
            # In case, the word is not in the dictionary; added it with one unit
            else:
                my_dict[i] = 1
#try and except blocks ensure that the user provide a valid file
try:
    introduction = """Kevin Mejia Bonilla, Peoplesoft ID: 1111111

    Program #7: Word Count: Dictionaries and Files
     
    Scientific Computing With Python
     
    Fall 2023

    This program computes the counting of each word in a file. """
    print(f"{introduction} \n")
    
    str_ = (sys.argv[1])
    fileHandle = open(str_)
    content = fileHandle.readlines()
    fileHandle.close()

    my_dict = {}
    for i in content:
        funct_1(i, my_dict)
    for k,v in my_dict.items():
        print(k,v)
except IndexError:
    print(f'Please provide a file name as an additional command line argument')
except FileNotFoundError as e:
    print(f"The file name you used {str_} does not exist. Please provide a file that exists.") 
```
